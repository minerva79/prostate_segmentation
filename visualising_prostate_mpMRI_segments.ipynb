# Visualising multiparametric Magnetic Resonance Imaging

Example of using NiBabel to visualise mpMRI prostate sections (http://medicaldecathlon.com/)

```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Load the NIfTI file
prostate_00 = nib.load("/content/drive/MyDrive/Task05_Prostate/imagesTr/prostate_00.nii")

# Extract the image data
prostate_00_data = prostate_00.get_fdata()

# Print the shape and data type of the image data
print(prostate_00_data.shape)
print(prostate_00_data.dtype)

# Assuming the last dimension is a modality/channel and we're interested in the first one
# Adjust the indexing to select the first modality/channel
prostate_00_data_modality = prostate_00_data[:, :, :, 0]

# Display the first 15 slices of the adjusted image data
fig, axes = plt.subplots(3, 5, figsize=(15, 9)) # Adjust subplot grid and size as needed
for i, ax in enumerate(axes.flat):
    if i < prostate_00_data_modality.shape[2]:  # Ensure we do not go out of bounds
        ax.imshow(prostate_00_data_modality[:, :, i])
        ax.axis('off')  # Hide axes for a cleaner look
    else:
        ax.axis('off')  # Hide axes if no more slices are available to display
plt.tight_layout()
plt.show()
```

```python
# Load the NIfTI file
labelled_prostate_00 = nib.load("/content/drive/MyDrive/Task05_Prostate/labelsTr/prostate_00.nii")

# Extract the image data
labelled_prostate_00_data = labelled_prostate_00.get_fdata()

# Print the shape and data type of the image data
print(labelled_prostate_00_data.shape)
print(labelled_prostate_00_data.dtype)

# The data is already in the correct format for visualization
# No need to adjust for modality/channel since it's a 3D dataset

# Display the first 15 slices of the image data
fig, axes = plt.subplots(3, 5, figsize=(15, 9))  # Adjust subplot grid and size as needed
for i, ax in enumerate(axes.flat):
    if i < labelled_prostate_00_data.shape[2]:  # Ensure we do not go out of bounds
        ax.imshow(labelled_prostate_00_data[:, :, i], cmap='gray')
        ax.axis('off')  # Hide axes for a cleaner look
    else:
        ax.axis('off')  # Hide axes if no more slices are available to display
plt.tight_layout()
plt.show()
```
